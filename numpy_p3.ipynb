# Secuencia de números

# Graficando funciones 2D/3D con $\texttt{np.meshgrid()}$ y Matplotlib

Con la ayuda de $\texttt{matplotlib}$ y el módulo $\texttt{np.meshgrid()}$ podemos realizar gráficos 2D y 3D.

## $\texttt{plt.contourf([x, y,] z)}$

Esta función toma tres argumentos: una malla de los valores $\texttt{x}$, una malla de los valores $\texttt{y}$, y una malla de los valores $\texttt{z}$. Los valores de $\texttt{x}$ y $\texttt{y}$ representan la posición en el dibujo y $\texttt{z}$ representa el nivel del contorno.

Graficar la función $z=(x^2+y^2)^{1/2}$, para $x\in[-2,2]$ y $y\in[-2,2]$

```python
x = np.linspace(-2,2)
y = np.linspace(-2,2)
```

```python
print(x.shape, y.shape)
```

```python
xx, yy = np.meshgrid(x,y,sparse=True)
```

```python
print(xx.shape, yy.shape)
```

```python
z = np.sqrt(xx**2 + yy**2)
```

```python
print(z.shape)
```

```python
h = plt.contourf(x, y, z, cmap="RdGy")

plt.axis('scaled')
plt.colorbar()
plt.show()
```

## $\texttt{plt.imshow(X, interpolation=None, extent=None)}$

En lugar de visualizar los niveles de contorno de manera discreta podemos hacerla de manera continua, pero para eso necesitamos la función $\texttt{plt.imshow()}$

```python
plt.style.use('default')
```

```python
print(z.shape)
```

```python
plt.imshow(z,extent=(x.min(), x.max(), y.min(), y.max()), cmap="RdGy")
plt.show()
```

Pondemos mejorar esta gráfica haciendo una interpolación.

```python
plt.imshow(z, interpolation='bicubic',extent=(x.min(), x.max(), y.min(), y.max()),cmap="RdGy")
plt.colorbar()
plt.show()
```

## Superficies

$z=e^{-(x^2+y^2)/2}$

```python
x = np.linspace(-5,5)
y = np.linspace(-5,5)

xx, yy = np.meshgrid(x, y, sparse=False)

z = np.exp(-(xx**2 + yy**2) / 2)
```

```python
plt.style.use('ggplot')

fig = plt.figure(figsize=(6,4))
ax  = fig.add_subplot(projection='3d')

ax.plot_surface(xx, yy, z, rstride=1, cstride=1, cmap='plasma')

plt.savefig("plot.png")
plt.show()
```

### Dibujando un Cono

* $x = r \cdot \cos{\theta}$
* $y = r \cdot \sin{\theta}$
* $z = r$

Donde $\theta\in[0,2\pi]$ y $r\in[0,h]$

```python
plt.style.use('ggplot')
```

```python
t = np.linspace(0, 2 * math.pi)
r = np.linspace(0, 1.5)
```

```python
# Creando la malla
tt, rr = np.meshgrid(t, r)
```

```python
x = rr * np.cos(tt)
y = rr * np.sin(tt)
z = 1.5 - rr
```

```python
fig = plt.figure()
ax  = fig.add_subplot(projection='3d')

ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap='viridis')

ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$')
ax.set_zlabel(r'$z$')
plt.show()
```

### Dibujando un Elipsoide

$\frac{x^2}{a^2}+\frac{y^2}{b^2}+\frac{z^2}{c^2}=1$

Solución:

Para graficar este tipo de superficies, primero tenemos que parametrizar $x,y,z$. Entonces los valores que satisfacen la ecuación del elipsoide son:

* $x=a\cos\theta\sin\phi$
* $y=b\sin\theta\sin\phi$
* $z=c\cos\phi$

Donde $\theta\in[0,2\pi]$ y $\phi\in[0,\pi]$

```python
a = 1
b = 2
c = 1

num_points = 100

theta = np.linspace(0, 2 * math.pi, num_points)
phi   = np.linspace(0, math.pi, num_points)

# Creando la malla
tt, pp = np.meshgrid(theta, phi)

# Calculando x,y,z
xx = a * np.cos(tt) * np.sin(pp)
yy = b * np.sin(tt) * np.sin(pp)
zz = c * np.cos(pp)
```

```python
fig = plt.figure()

ax = plt.axes(projection='3d')
ax.plot_surface(xx, yy, zz, rstride=1, cstride=1, cmap='inferno')

ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$')
ax.set_zlabel(r'$z$')

ax.set_xlim(-2,2)
ax.set_ylim(-4,4)
ax.set_zlim(-2,2)

ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])

ax.view_init(elev=35, azim=120)
plt.show()
```
