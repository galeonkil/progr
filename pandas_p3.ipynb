## Dataset: Air Quality (NO2 diffusion tube) data

Los datos que se muestran son valores de dióxido de nitrógeno (NO2) para las estaciones de medición FR04014, BETR801 y London Westminster en París, Amberes y Londres, respectivamente

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/air_quality_no2.csv")
```

```python
df.head(10)
```

```python
df["station_london"].plot(figsize=(16, 6))
```

Puedo también controlar el color de la gráfica

```python
df["station_london"].plot.area(figsize=(16, 6), color={'station_london':'red'})
```

```python
df.plot.area(figsize=(16, 6))
```

```python
df.plot(figsize=(16, 6))
```

# Scatter

```python
df.plot.scatter(x="station_antwerp", y="station_paris", alpha=0.5, figsize=(6, 6))
```

```python
df.plot.scatter(x="station_antwerp", y="station_london", alpha=1, figsize=(9, 6))
```

Puedo también controlar el color de los puntos

```python
df.plot.scatter(x="station_antwerp", y="station_london", c="#519430", alpha=.8, figsize=(9, 6))
```
