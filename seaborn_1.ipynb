```python
import seaborn as sns

sns.set_theme(style="darkgrid")
```

# Relacionando variables con **gráficos de dispersión** (scatter plots)

Este tipo de visualización representa la distribución conjunta de dos variables, usando una nube de puntos. Donde cada punto representa una observación en la base de datos. Para esto, usamos $\texttt{$\texttt{scatterplot()}$}$ en Seaborn.

## Dataset "tips"

En esta dataset hay información sobre propinas dejadas en un restaurante, con información sobre el día en el que se realizó, si fue en el almuerzo o la cena, el número de comensales, el sexo de la persona que dejó la propina, si era fumador o no, etc.

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
# @title total_bill vs tip

from matplotlib import pyplot as plt
tips.plot(kind='scatter', x='total_bill', y='tip', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
sns.relplot(x="total_bill", y="tip", kind="scatter", data=tips)
```

Por defecto $\texttt{sns.relplot()}$ realiza un gráfico de disperción con  $\texttt{kind="scatter"}$.

**Pregunta: ¿Que podemos decir de la gráfica?**

También podemos otra dimensión a nuestra gráfica coloreando los puntos de acuerdo a esta nueva variable, añadiendo si la persona era fumador o no.

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips)
```

Para realizar una mejor visualización podemos cambiar el estilo segun la columna *smoker*.

```python
sns.relplot(x="total_bill", y="tip", style="smoker", data=tips)
```

El estilo puede ser representado por otra variable más.

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", style="time", data=tips)
```

La colorización también se puede dar con respecto a variables con valores numéricos, creando una paleta de colores.

```python
tips.head()
```

```python
sns.relplot(x="total_bill", y="tip", hue="size", data=tips)
```

La paleta de colores se puede editar: [Paleta de colores](https://seaborn.pydata.org/tutorial/color_palettes.html)

```python
sns.relplot(x="total_bill", y="tip", hue="size", palette="ch:r=-0.5,l=0.9", data=tips)
```

En lugar de cambiar el color de los puntos de acuerdo a la varibale *size*, podemos cambinar el tamaño de los puntos.

```python
sns.relplot(x="total_bill", y="tip", size="size", data=tips)
```

El rango del tamaño de los puntos se puede editar tambien.

```python
sns.relplot(x="total_bill", y="tip", size="size", sizes=(10, 200), data=tips);
```
