# Visualizando variables categóricas

Si alguna de las variables es *categorica* (valores discretos), es necesario otra forma de visualización. Para este proceso utilizaremos $\texttt{catplot()}$.

```python
import seaborn as sns
```

Dataset: **tips**

```python
tips = sns.load_dataset("tips")
tips.head()
```

Por defecto $\texttt{catplot()}$ realiza un gráfico de dispersión.

```python
sns.catplot(x="day", y="total_bill", data=tips)
```

Podemos controlar la oscilación.

```python
sns.catplot(x="day", y="total_bill", jitter=False, data=tips)
```

Podemos mejorar esta visualización con $\texttt{kind="swarm"}$, de tal modo que no exista superposición entre cada punto.

```python
sns.catplot(x="day", y="total_bill", kind="swarm", data=tips)
```

En este caso, también podemos añadir una dimensión más, añadiendo color a nuestro gráfico.

```python
sns.catplot(x="day", y="total_bill", hue="sex", kind="swarm", data=tips)
```

Si colocamos la variable $\texttt{x}$ con una variable categórica numéricam, automáticamente Seaborn ordenará los datos.

```python
sns.catplot(x="size", y="total_bill", data=tips)
```

Sin embargo, este orden puede ser controlado por una variable.

```python
sns.catplot(x="size", y="total_bill", order=[6,5,4,3,2,1], data=tips)
```

```python
sns.catplot(x="smoker", y="tip", order=["Yes", "No"], data=tips)
```

También podemos colocar la variable categórica en la ordenada.

```python
sns.catplot(x="total_bill", y="day", hue="time", kind="swarm", data=tips)
```

## BoxPlot

Conforme el tamaño de la base de datos crece, la gráfica de disperción puede no ser lo sufience descriptiva. Para este caso, podemos utilizar un BoxPlot (Diagrama de caja). En Seaborn podemos realizar estas gráficas con $\texttt{kind=box}$.

```python
sns.catplot(x="day", y="total_bill", kind="box", data=tips)
```

Del mismo modo, podemos añadir una variable más a nuestra gráfica.

```python
sns.catplot(x="day", y="total_bill", hue="smoker", kind="box", data=tips)
```

Esta nueva variable podría aplicarse directamente a la variable categórica (como un filtro). Para esto, vamos a crear primero una columna que representa a el fin de semana.

```python
tips["finde"] = tips["day"].isin(["Sat", "Sun"])
```

```python
tips["total_bill"].median()
```

```python
sns.catplot(x="day", y="total_bill", hue="finde", kind="box", data=tips)
```

Por defecto, cuando utilizamos una variable adicional, se reduce el ancho de la cajas. Sin embargo, nostros podemos cambiar esto manualmente.

```python
sns.catplot(x="day", y="total_bill", hue="finde",
            kind="box", dodge=True, data=tips)
```
